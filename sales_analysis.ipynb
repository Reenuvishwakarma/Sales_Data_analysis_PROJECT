```python
import numpy as numpy
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set_style("whitegrid")
```

```python
data = {
    "Product": ["Laptop","Phone","Tablet","Laptop","Phone","Watch","Tablet","Phone"],
    "Category": ["Electronics","Electronics","Electronics","Electronics","Electronics","Accessories","Electronics","Electronics"],
    "Region": ["North","South","East","West","North","South","West","East"],
    "Sales": [80000,50000,30000,75000,45000,15000,35000,55000],
    "Quantity": [5,8,4,6,7,10,5,9],
    "Profit": [12000,8000,5000,10000,7000,3000,6000,9000]
}
```

```python
df=pd.DataFrame(data)
print(df)
```

```python
print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
print(df.shape)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.columns
```

```python
df.groupby("Product")["Sales"].sum()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(
    data=df,
    x="Product",
    y="Sales",
    estimator=sum
)
plt.title("Product wise Sales")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```

```python
df.groupby("Region")["Sales"].sum()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(
    data=df,
    x="Region",
    y="Sales",
    estimator=sum
)
plt.title("Region wise Sales")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x="Sales",
    y="Profit",
    hue="Product",
    s=100
)

plt.title("Sales vs Profit Relationship")

plt.xlabel("Sales")
plt.ylabel("Profit")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    y="Sales"
)

plt.title("Sales Distribution and Outliers")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x="Product",
    y="Sales"
)

plt.title("Product Wise Sales Distribution")

plt.show()
```

```python
corr = df.corr(numeric_only=True)

corr
```

```python
plt.figure(figsize=(8,5))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.show()
```
